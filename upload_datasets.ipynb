```python
dataset_ids = [46773, 46774, 46775, 46776, 46777, 46778, 46779, 46780, 46781, 46782, 46783]
task_ids = []
for dataset_id in dataset_ids:
    dataset = openml.datasets.get_dataset(dataset_id, download_qualities=True)
    if dataset.qualities['NumberOfClasses'] > 0:
        task_type = openml.tasks.TaskType.SUPERVISED_CLASSIFICATION
        estimation_procedure_id = 1  # '10-fold Crossvalidation'
    else:
        task_type = openml.tasks.TaskType.SUPERVISED_REGRESSION
        estimation_procedure_id = 7  # '10-fold Crossvalidation'

    task = openml.tasks.create_task(task_type=task_type, dataset_id = dataset.id, estimation_procedure_id=estimation_procedure_id, target_name=dataset.default_target_attribute)
    task.publish()
    task_id = task.id
    task_ids.append(task_id)
```
