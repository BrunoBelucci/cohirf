```python
df = X.copy()
df["target"] = y
# sub-sample with weights
weights = df["target"].value_counts(normalize=True)
df["weights"] = df["target"].map(weights)
df = df.sample(n=100000, random_state=42, weights=df["weights"])
df = df.drop(columns=["weights"])
X = df.drop(columns=["target"])
y = df["target"]
```

```python
save_path = Path("/home/belucci/code/cohirf/data/kdd-99-sub")
X.to_csv(save_path / "X.csv")
y.to_csv(save_path / "y.csv")
```
