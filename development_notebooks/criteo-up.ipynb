```python
import pandas as pd
from pathlib import Path
```

```python
path = Path("/home/belucci/downloads/criteo-research-uplift-v2.1.csv")
df = pd.read_csv(path)
```

```python
df['label'] = df['conversion'].astype(str) + df['visit'].astype(str) + df['exposure'].astype(str)
```

```python
df["label"] = df["label"].astype("category")
```

```python
df["label"].cat.categories
```

```python
df["label"].dtype
```

```python
df['label'] = df['label'].astype('category').cat.codes
```

```python
df = df[~df.duplicated()]
df = df.drop(columns=['conversion', 'visit', 'exposure'])
```

```python
X = df.drop(columns=['label'])
y = df['label']
```

```python
X.to_csv("/home/belucci/code/cohirf/data/criteo-up/X.csv")
y.to_csv("/home/belucci/code/cohirf/data/criteo-up/y.csv")
```

```python
# Sub sample of criteo, ~1 M, all samples from labels 1 - 3 + 500_000 samples from label 0 (4 and 5 are too few)
df_sampled = df.loc[~df['label'].isin([0, 5, 4])].copy()
df_sampled = pd.concat([df_sampled, df.loc[df['label'] == 0].sample(n=500_000, random_state=42)], axis=0)


# weights = df['label'].value_counts(normalize=True)
# df['weights'] = df['label'].map(weights)
# df = df.sample(n=100000, random_state=42, weights=df['weights'])
# df = df.drop(columns=['weights'])
# X = df.drop(columns=["label"])
# y = df["label"]
```
