```python
import pandas as pd
import openml
```

```python
df = pd.read_csv('characteristics.csv')
```

```python
datasets_ids = [40685, 39, 61, 182, 40984, 1478, 1568]
datasets_ids = [46773, 46779, 46783]
df.loc[df['dataset_id'].isin(datasets_ids)]
```

```python
task_ids = [145977, 59, 2074, 14970, 9892, 146212, 146822, 363483, 363484, 363485, 363486, 363487, 363488, 363489, 363490, 363491, 363492, 363493]
df = []
for task in task_ids:
    task = openml.tasks.get_task(task)
    dataset = openml.datasets.get_dataset(task.dataset_id)
    characteristics = {
        'task_id': task.id,
        'dataset_id': dataset.id,
        'name': dataset.name,
        'task': task.task_type,
        'evaluation': task.estimation_procedure,
        'n_instances': dataset.qualities['NumberOfInstances'],
        'n_features': dataset.qualities['NumberOfFeatures'],
        'n_categorical_features': dataset.qualities['NumberOfSymbolicFeatures'],
        'n_continuous_features': dataset.qualities['NumberOfNumericFeatures'],
        'n_nans': dataset.qualities['NumberOfMissingValues'],
        'n_rows_nans': dataset.qualities['NumberOfInstancesWithMissingValues'],
        'n_classes': dataset.qualities['NumberOfClasses'],
        'url': dataset.openml_url,
    }
    df.append(characteristics)

df = pd.DataFrame.from_dict(df)
```
